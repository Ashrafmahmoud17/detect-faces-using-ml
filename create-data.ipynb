```python
import numpy as np
import cv2
import mediapipe as mp
import time


mp_face_mesh = mp.solutions.face_mesh
mp_drawing = mp.solutions.drawing_utils

face_mesh = mp_face_mesh.FaceMesh(
    max_num_faces=1,
    refine_landmarks=True,
    min_detection_confidence=0.5,
    min_tracking_confidence=0.5
)

cap=cv2.VideoCapture(0)
skip = 0
Ashraf = []
while True:
    ret,image=cap.read()
    if ret==False :
        break
    rows, cols, _ = image.shape
    if skip % 5 == 0:
        image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
        image.flags.writeable = False
        results = face_mesh.process(image)
        image.flags.writeable = True
        
        if results.multi_face_landmarks:
            for face_landmarks in results.multi_face_landmarks:
                swap = []
                x = [ int(landmark.x * cols) for landmark in face_landmarks.landmark]
                y = [ int(landmark.y * rows) for landmark in face_landmarks.landmark]
                swap.extend(x)
                swap.extend(y)
                Ashraf.append(swap)
                mp_drawing.draw_landmarks( image, face_landmarks,mp_face_mesh.FACEMESH_TESSELATION,mp_drawing.DrawingSpec(color=(0,255,0), thickness=1, circle_radius=1),mp_drawing.DrawingSpec(color=(0,0,255), thickness=1))
        image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)
    skip = skip + 1
    
    
    

    cv2.imshow('MediaPipe Faces', image)
    if cv2.waitKey(1) & 0xFF == ord("q"):
        break
cap.release()
cv2.destroyAllWindows()
```

```python
import numpy as np
import cv2
import mediapipe as mp
import os

mp_face_mesh = mp.solutions.face_mesh

face_mesh = mp_face_mesh.FaceMesh(
    max_num_faces=1,
    refine_landmarks=True,
    min_detection_confidence=0.5,
    min_tracking_confidence=0.5
)

cap = cv2.VideoCapture(0)

count = 0

save_folder = "saved_faces"

os.makedirs(save_folder, exist_ok=True)

while True:
    ret, image = cap.read()

    if not ret:
        break

    rows, cols, _ = image.shape

    # RGB
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    results = face_mesh.process(image_rgb)

    if results.multi_face_landmarks:

        for face_landmarks in results.multi_face_landmarks:

            swap = []

            x = [int(lm.x * cols) for lm in face_landmarks.landmark]
            y = [int(lm.y * rows) for lm in face_landmarks.landmark]

            swap.extend(x)
            swap.extend(y)

            swap = np.array(swap)

            swap = scaler.transform(swap.reshape(1, -1))

            y_pred = model.predict(swap)[0]

            cv2.putText(
                image,
                str(y_pred),
                (10, 50),
                cv2.FONT_HERSHEY_SIMPLEX,
                1.2,
                (255, 0, 0),
                3
            )

            if count % 20 == 0:

                filename = f"{save_folder}/{y_pred}_{count}.jpg"

                cv2.imwrite(filename, image)

                print(f"Saved: {filename}")

            count += 1

    cv2.imshow("Face Prediction", image)

    if cv2.waitKey(1) & 0xFF == ord("q"):
        break

cap.release()
cv2.destroyAllWindows()
```
